```python
import torch

ckpt_path = "/content/drive/MyDrive/Mae/data/VIT_B_16x4_MAE_PT.pyth"  #path to ckpt
ckpt = torch.load(ckpt_path, map_location="cpu")

state = ckpt.get("model_state", ckpt.get("state_dict", ckpt))

def fuse_qkv(sd, prefix):
    """
    prefix example: 'pred_head.transforms.0.0.attn.'
    converts:
      q_proj.weight, k_proj.weight, v_proj.weight -> qkv.weight (cat on dim=0)
      q_proj.bias,   k_proj.bias,   v_proj.bias   -> qkv.bias   (cat on dim=0)
    """
    q_w = sd.pop(prefix + "q_proj.weight", None)
    k_w = sd.pop(prefix + "k_proj.weight", None)
    v_w = sd.pop(prefix + "v_proj.weight", None)
    q_b = sd.pop(prefix + "q_proj.bias",   None)
    k_b = sd.pop(prefix + "k_proj.bias",   None)
    v_b = sd.pop(prefix + "v_proj.bias",   None)

    if q_w is not None and k_w is not None and v_w is not None:
        sd[prefix + "qkv.weight"] = torch.cat([q_w, k_w, v_w], dim=0)
    if q_b is not None and k_b is not None and v_b is not None:
        sd[prefix + "qkv.bias"] = torch.cat([q_b, k_b, v_b], dim=0)

for layer in [0,1,2,3]:
    for sub in [0,1]:
        fuse_qkv(state, f"pred_head.transforms.{layer}.{sub}.attn.")


torch.save({"model_state": state}, "/content/drive/MyDrive/Mae/data/VIT_B_16x4_MAE_PT_FUSED.pyth")
print("Wrote fused checkpoint.")
```

```python
import torch, re

CKPT_IN  = "/content/drive/MyDrive/Mae/data/VIT_B_16x4_MAE_PT.pyth"
CKPT_OUT = "/content/drive/MyDrive/Mae/data/VIT_B_16x4_MAE_PT_FUSED.pyth"

ckpt = torch.load(CKPT_IN, map_location="cpu")
sd = ckpt.get("model_state", ckpt.get("state_dict", ckpt))

# Find all decoder attention prefixes that use separate q/k/v
# e.g., 'pred_head.transforms.0.0.attn.'
q_keys = [k for k in sd.keys() if re.match(r"^pred_head\.transforms\.\d+\.\d+\.attn\.q\.(weight|bias)$", k)]
prefixes = sorted(set(k.rsplit("q.", 1)[0] for k in q_keys))

fused_count = 0
for pref in prefixes:
    q_w = sd.pop(pref + "q.weight", None); q_b = sd.pop(pref + "q.bias", None)
    k_w = sd.pop(pref + "k.weight", None); k_b = sd.pop(pref + "k.bias", None)
    v_w = sd.pop(pref + "v.weight", None); v_b = sd.pop(pref + "v.bias", None)

    # require all three weights to exist
    if q_w is None or k_w is None or v_w is None:
        # put back anything we popped (safety)
        if q_w is not None: sd[pref + "q.weight"] = q_w
        if k_w is not None: sd[pref + "k.weight"] = k_w
        if v_w is not None: sd[pref + "v.weight"] = v_w
        if q_b is not None: sd[pref + "q.bias"]   = q_b
        if k_b is not None: sd[pref + "k.bias"]   = k_b
        if v_b is not None: sd[pref + "v.bias"]   = v_b
        continue

    sd[pref + "qkv.weight"] = torch.cat([q_w, k_w, v_w], dim=0)
    if q_b is not None and k_b is not None and v_b is not None:
        sd[pref + "qkv.bias"] = torch.cat([q_b, k_b, v_b], dim=0)

    fused_count += 1

print("Fused decoder attn blocks:", fused_count)


cand_proj = [k for k in sd.keys() if k.startswith("pred_head.") and "proj" in k and k.endswith(".weight")]
print("Projection-like keys:", cand_proj[:5])


torch.save({"model_state": sd, "cfg": ckpt.get("cfg")}, CKPT_OUT)
print("Saved:", CKPT_OUT)
```
